```python
#importing the libraries for our model for the prediction
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df["embarked"].fillna("s",inplace=True)
```

```python
df['deck'].fillna("unknown",inplace=True)
```

```python
df.drop(["class","who","adult_male","deck","embark_town","alive"],axis=1,inplace=True)
```

```python
df["sex"]=df["sex"].map({"male":0,"female":1})
df["embarked"]=df["embarked"].map({"S":0,"C":1,"Q":2})
df["alone"]=df["alone"].map({True:1,False:0})
df.fillna(0,inplace=True)
```

```python
#Feature selection
#step:to get input and output extracting the features:
# x= we will take always for input and y=output this standard way for get proper input and output
x=df.drop("survived",axis=1)# drop to remove the survived column,except survived everything we will input ,axis1 means rowwise and 0 means columns wise
y=df["survived"]# survived will become the output
```

```python
#train and test=>test size means 20%of data we are taking,randow_state its 42 rows and after that randomly its take data for test and train
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42
)
```

```python
#Model Building using Logistic regresssion
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
#predict the output
y_pred=model.predict(x_test)
```

```python
#Evaluate the model to check the accuracy and comfusion matrix
print("Accuracy:",accuracy_score(y_test,y_pred))#if run we get different accuracy bcz of train and test evry model take differntly input so
```

```python
#confusion matrix
cm=confusion_matrix(y_test,y_pred)#logic to get confusion matrix
sns.heatmap(cm,annot=True,fmt="d")#fmd its show in the form integer
plt.title("confusion matrix")#
plt.show()
```
